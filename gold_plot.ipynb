```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt_realtime
import pandas as pd
import numpy as np 

#Open csv file for gold prices
gold_df = pd.read_csv('Gold_Silver_Analysis/gold_prices.csv')
```

```python
gold_df[::-1]
```

```python
#Title for gold selling prices
plt.title("Gold selling price index")

#Names for x and y axis
plt.xlabel('Date')
plt.ylabel('Price')

#Plot the x and y into a graph
gold_plot = plt.plot(gold_df[::-1]["Date"], gold_df[::-1]["Close/Last"])

#Show the plot
plt.show(gold_plot)

#Save figure of the plot
plt.savefig("gold_plot.png")
```

We can see that gold started at 
$1365.7 on 08/19/2013.
After a downfall and it returned back to the price at
